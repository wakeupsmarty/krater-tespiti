```python
import yaml

data = {
    'train': '/content/dataset/LU3M6TGT_yolo_format/train/images',
    'val': '/content/dataset/LU3M6TGT_yolo_format/valid/images',
    'nc': 1,
    'names': ['crater']
}

with open('/content/dataset/LU3M6TGT_yolo_format/data.yaml', 'w') as f:
    yaml.dump(data, f)

print("data.yaml güncellendi!")
!cat /content/dataset/LU3M6TGT_yolo_format/data.yaml
```

```python
from ultralytics import YOLO

model = YOLO('yolov8n.pt')  # nano model, hızlı

model.train(
    data='/content/dataset/LU3M6TGT_yolo_format/data.yaml',
    epochs=50,
    imgsz=640,
    batch=16,
    name='crater_detection'
)
```

```python
# Sonuç grafiklerini göster
from IPython.display import Image
Image('/content/runs/detect/crater_detection/results.png')
```

```python
# Test tahmini yap
from ultralytics import YOLO
model = YOLO('/content/runs/detect/crater_detection/weights/best.pt')

# Validation setinden birkaç görsel üzerinde tahmin
results = model.predict(
    source='/content/dataset/LU3M6TGT_yolo_format/valid/images',
    conf=0.25,
    save=True,
    max_det=100
)
```
